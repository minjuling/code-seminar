```python
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
image1 = Image.open('../data_semantics/training/image_2/000000_10.png')
image2 = Image.open('../data_semantics/training/image_2/000001_10.png')
```

```python
plt.imshow(image1)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image1)
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))


ax1 = fig.add_subplot(3, 1, 1)
ax1.imshow(image1)
ax1.set_title('Image')

ax2 = fig.add_subplot(3, 1, 2)
ax2.imshow(image2)
ax2.set_title('Image')

ax2 = fig.add_subplot(3, 1, 3)
ax2.imshow(image1)
ax2.set_title('Image')
plt.show()
```

```python
color_img = cv2.imread('../data_semantics/training/image_2/000000_10.png', cv2.IMREAD_COLOR)
gray_img = cv2.imread('../data_semantics/training/image_2/000000_10.png', cv2.IMREAD_GRAYSCALE)
unch_img = cv2.imread('../data_semantics/training/image_2/000000_10.png', cv2.IMREAD_UNCHANGED)
```

```python
cv2.imshow('color_img', color_img)
cv2.imshow('gray_img', gray_img)
cv2.imshow('unch_img', unch_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
h_concat = cv2.hconcat([color_img, unch_img])
print(h_concat.shape)

# h_concat = cv2.hconcat([h_concat, gray_img])
cv2.imshow('h_concat', h_concat)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
cv2.imwrite('../out/Gray.png', gray_img)
cv2.imwrite('../out/color.png', color_img)
cv2.imwrite('../out/unch.png', unch_img)
```
