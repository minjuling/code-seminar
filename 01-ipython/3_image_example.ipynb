# Image

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
from PIL import Image
import numpy as np
```

```python
image_left = Image.open('../dataset/image_0_left.png')
image_right = Image.open('../dataset/image_0_right.png')
```

```python
plt.imshow(image_left)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.imshow(image_left)
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(image_left)
ax1.set_title('left Image')

ax2 = fig.add_subplot(rows, cols, 2)
ax2.imshow(image_right)
ax2.set_title('right Image')

plt.show()
```

```python
color_img = cv2.imread('../dataset/image_0_left.png', cv2.IMREAD_COLOR)
gray_img = cv2.imread('../dataset/image_0_left.png', cv2.IMREAD_GRAYSCALE)
unch_img = cv2.imread('../dataset/image_0_left.png', cv2.IMREAD_UNCHANGED)

```

```python
cv2.imshow('color_img', color_img)
cv2.imshow('gray_img', gray_img)
cv2.imshow('unch_img', unch_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
h_concat = cv2.hconcat([color_img, unch_img])
print(h_concat.shape)

# h_concat = cv2.hconcat([h_concat, gray_img])
cv2.imshow('h_concat', h_concat)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
cv2.imwrite('../out/Gray.png', gray_img)
cv2.imwrite('../out/color.png', color_img)
cv2.imwrite('../out/unch.png', unch_img)
```
